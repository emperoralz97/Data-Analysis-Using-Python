## Plotly Imports

```python
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot
init_notebook_mode(connected=True) 
```

**Import pandas and read the csv file: 2014_World_Power_Consumption**

```python
import pandas as pd
import numpy as np
```

```python
try:
    df = pd.read_csv('2014_World_Power_Consumption.csv')
    print("File Loaded successfully...")
except:
    print("File not found...")
```

**Check the head of the DataFrame.**

```python
df.head()
```

```python
def clear_comma(country):
    if ',' in country:
        return country[-1]
    else:
        return country
df.Country = df.Country.apply(clear_comma)
df.drop(df.loc[df.Country==','].index,inplace =True)
```

**Referencing the lecture notes, create a Choropleth Plot of the Power Consumption for Countries using the data and layout dictionary.**

```python
# 1. Creating the data dictionary
data = dict(type = 'choropleth',
           text = df.Country,
            locations=df.Country,
            locationmode='country names',
            colorscale = 'Blackbody',
            z=df['Power Consumption KWH'],
            colorbar=dict(title='Consumption of Power KWH')
           )
layout = dict(
    title = '2014 Power Consumption for Countries',
    geo = dict(
        showframe = True,
        projection = {'type':'natural earth'}
    )
)
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```

## USA Choropleth

**Import the 2012_Election_Data csv file using pandas.**

```python
try:
    election = pd.read_csv('2012_Election_Data')
    print('File loaded successfully...')
except:
    print('File not loaded successfully.')
```

**Check the head of the DataFrame.**

```python
election.head()
```

**Now create a plot that displays the Voting-Age Population (VAP) per state. If you later want to play around with other columns, make sure you consider their data type. VAP has already been transformed to a float for you.**

```python
# Drawing a plot for VAP per state for united states map
```

```python
# Creating the data and layout variable
data = dict(type = 'choropleth',
           text = election['State'],
           locations = election['State Abv'],
           locationmode = 'USA-states',
           colorscale='Picnic',
            z = election['Voting-Age Population (VAP)'],
            colorbar = dict(title='Voting-Age Population(VAP)')
           )
layout = dict(title = '2014 Voting-Age Population (VAP) per state',
             geo = dict(scope = 'usa'))
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```
