```python
import plotly.plotly as py
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode, plot,iplot
```

```python
init_notebook_mode(connected=True)
```

```python
import plotly.graph_objs as go
```

**Create two variable: 1.data 2.layout**

```python
# Build the data dictionary
# Refer documentation of plotly
data = dict(type = 'choropleth',
           locations = ['AZ','CA','NY'],
           locationmode = 'USA-states',
           colorscale='Jet',
           text = ['Arizona','Cali','New York'],
           z = [1.0,2.0,3.0],# Value you represent as color
           colorbar= dict(title = 'Colorbal Title'))
```

```python
layout = dict(geo=dict(scope = 'usa'))
```

```python
choromap = go.Figure(data=[data],layout=layout)
```

```python
iplot(choromap)
```

```python
import pandas as pd
```

```python
try:
    df = pd.read_csv('2011_US_AGRI_Exports.csv')
    print('File loaded succesfully')
except:
    print('File not found...')
```

```python
df.head()
```

```python
# 1. Create the data variable
data = dict(type='choropleth',
            colorscale = 'Greens',
            locations = df['code'],
            z = df['total exports'],
            locationmode = 'USA-states',
            text = df['text'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 1)),
            colorbar = {'title':"Millions USD"}
            ) 
```

```python
# 2. Create the layout variable
layout = dict(title = '2011 US Agriculture Exports by State',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)')
             )
```

```python
choromap2 = go.Figure(data = [data],layout = layout)
```

```python
iplot(choromap2)
```

# Part-2 Internation Map

```python
df2 = pd.read_csv('2014_World_GDP.csv')
```

```python
df2.head()
```

```python
data = dict(type='choropleth',
           locations = df2.CODE,
           z = df2['GDP (BILLIONS)'],
           text = df2.COUNTRY,
           colorbar = dict(title='GDP in Billion USD'))
```

```python
layout = dict(title = '2014 Global GDP', geo = dict(showframe = False,
                                                   projection = {'type':'natural earth'}))
```

```python
choromap3 = go.Figure([data],layout)
```

```python
iplot(choromap3)
```
