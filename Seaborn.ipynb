# Seaborn

## 1. Distribution Plots

```python
 # Importing module
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# DistPlot allows us to show distribution of univariable in the plot(one variable)
sns.distplot(tips.total_bill, kde = True)
plt.grid(True)
```

```python
# JointPlot combine distribution of two variable
# Default kind is scatter
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
# JointPlot combine distribution of two variable
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# JointPlot combine distribution of two variable
# Seaborn draws the regression line
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# JointPlot combine distribution of two variable
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# Pairplot plots pairwise relationship across the entire dataframe
# ie, it does joinplot for every numerical column in the dataframe
# hue : add column name containing categorical values
# palette: add colur map to your plot
sns.pairplot(tips,hue='sex',palette='Set1')
```

```python
# kde: Kernel density estimation = sum of all the normal distribution around a rugplot
# rugplot
sns.rugplot(tips['total_bill'])
```

```python
# Kde plot without any distribution of bars
sns.kdeplot(tips['total_bill'])
```

## 2. Categorical Plots

```python
import numpy as np
```

```python
# Focus more on categorical column 
# Barplot: agg of total value in y
# x: categorical
# y: numerical (mean)
# estimator: default is mean (used for statistical function)
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# Focus more on categorical column 
# 1.Barplot: agg of total value in y
# x: categorical
# y: numerical (mean)
# estimator: default is mean (used for statistical function)
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# 2. countplot: count the no. of occurences
# only set the x values
sns.countplot(x='sex',data=tips)
```

```python
# 3. Boxplot: Distribution of data for facilating variable
# x: categorical
# y: numerical
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# 3. Boxplot: Distribution of data across some category
# x: categorical
# y: numerical
# hue: categorical
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# 4. violinplot: Same as boxplot extra info 
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# 4. violinplot: Same as boxplot extra info 
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# 4. violinplot: Same as boxplot extra info 
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
# 5. stripplot: Draws a scatter plot where one var is categorical
# jitter: add random noise on normal plot
sns.stripplot(x='day',y='total_bill',data=tips)
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# 6. swarmplot: Combine scatter+strip+violin
# Cons: slow for large dataset
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# 6. swarmplot: Combine scatter+strip+violin
# Cons: slow for large dataset
# to show more info run it along with violinplot
# more useful for explotary data analysis than presentation
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# factorplot: same as the above along with kind{bar, violin, etc} more of a general plot
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

## 3. Matrix Plots

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
#Matrix form has both rows and column as variable
fc = flights.corr()
fc
```

```python
tc = tips.corr()
tc
```

```python
# Drawing the heatmap
sns.heatmap(tc)
```

```python
# 1. heatmap: Type of Matrixplot
sns.heatmap(tc,annot=True)
```

```python
# Drawing the heatmap
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
# Convert one of the columns to index
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths=3)
```

* **Popular month of flight is june july and august(summer month)**
* **As the year progressed no. of flight increased**

```python
# 2. Cluster plot: See the similarity of clusters based of the colouring
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

* **Nov, January, February are low passenger month**
* **July, August are high passenger month**

## 4. Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
 # Use PairGrid to have more control over the plotting compare to pairplot
g = sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# FactGrid: offers row and column on which you want to separate by
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
# FactGrid: offers row and column on which you want to separate by
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')# dealing with plot type having more than one argument
```

## 5. Regression Plot

```python
# 1. Linear Model plot
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# 1. Linear Model plot
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':50})
```

* **Male and Female shows the same trend with respect to tip vs total_bill**

```python
# 1. Linear Model plot
# Instead of separating by color separate by column
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```
