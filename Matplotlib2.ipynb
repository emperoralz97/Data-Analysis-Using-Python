```python
# dealing with matplotlib
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,50)
x
```

```python
sinus = np.sin(x)
sinus
```

```python
plt.plot(x, sinus)
plt.show()
```

```python
plt.plot(x,sinus,"o")
plt.show()
```

```python
# rapid multiplot
cosinus = np.cos(x)
plt.plot(x,sinus,"-b",x,sinus,"ob",x,cosinus,"-r",x,cosinus,"or")
plt.xlabel("this is x-axes")
plt.ylabel("this is y-axes")
plt.title("My First Plot")
plt.show()
```

```python
# step by step
plt.plot(x,sinus,label="sinus",color="blue",linestyle="--",linewidth=2)
plt.plot(x,cosinus,label="cosinus",color="red",linestyle="-",linewidth=2)
plt.legend()
plt.show()
```

```python
# printing scatter diagram
import pandas as pd
try:
    salary = pd.read_csv("e:\salary_table.csv")
except:
    url = "http://...."
    salary = pd.read_csv(url)
df = salary
df
```

```python
colors = colors_edu = {"Bachelor":"r","Master":"g","Ph.D":"blue"}
plt.scatter(df["experience"],df["salary"],c=df["education"].apply
            (lambda x:colors[x]),s=50)
plt.show()
```

```python
# figure size
plt.figure(figsize=(6,5))
# define colors and symbols manually
symbols_manage = dict(Y="*",N=".")
colors_edu = {"Bachelor":"r","Master":"g","Ph.D":"blue"}
# group by education and management thus will produce 6 groups
for values, d in salary.groupby(["education","management"]):
    edu,manager = values
    plt.scatter(d["experience"],d["salary"],marker=
                symbols_manage[manager],color=colors_edu[edu],
                s=150,label=manager+" / "+edu)
# set labels
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend(loc=4)       # i.e. lower right corner
plt.show()
```
